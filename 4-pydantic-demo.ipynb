### Pydantic Demo

First we define a class and pass the pydantic's base model as argument to it, so that it will inherit all its properties

```python
from pydantic import BaseModel
from typing import Optional

class Student(BaseModel):
    name: str = 'default-name-rocklee' #Here we can give default name incase while creating the obejct we dont pass the name
    age: Optional[int] = None #Here incase we dont want to give default value as well as make it optional
```

Simple case

```python
new_student = {'name': "naruto"} #this works
student = Student(**new_student)
print(student)
```

The case which does not execute because pydantic does its work

```python
new_student = {'name': 32} #this will throw error
student = Student(**new_student)
print(student)
```

Here, we pass the optional argument but not main name because it already has default value assigned to it

```python
new_student = {'age': 32} 
student= Student(**new_student)
print(student)
```

Pydantic is smart enough to perform type conversion even if we pass the wrong type , but right values.

For example, we know that age is int, so we passed 32 right, but what if we by mistake pass it as '32' which is string format.
Here pydantic automatically converts that value to string, because its possible.

```python
new_student = {'age': "32"} 
student= Student(**new_student)
print(student)
```

From pydantic we can even validate emails with the help of "EmailStr" import

```python
from pydantic import BaseModel, EmailStr
from typing import Optional

class Student(BaseModel):
    name: str = 'default-name-rocklee' #Here we can give default name incase while creating the obejct we dont pass the name
    age: Optional[int] = None #Here incase we dont want to give default value as well as make it optional
    email: EmailStr #This datatype itself is email string
```

Here, we pass a simple string to see whether it validates email or not, it should throw an error

```python
new_student = {'age': "32", "email": 'abc'} 
student= Student(**new_student)
print(student)
```

Now , we will pass a valid email address


```python
new_student = {'age': "32", "email": 'abc@outlook.com'} 
student= Student(**new_student)
print(student)
```

We can even set the field values like we can set the range of limits of a number like (0 to 10), it will even enforce that.

```python
from pydantic import BaseModel, EmailStr, Field
from typing import Optional

class Student(BaseModel):
    name: str = 'default-name-rocklee' #Here we can give default name incase while creating the obejct we dont pass the name
    age: Optional[int] = None #Here incase we dont want to give default value as well as make it optional
    email: EmailStr #This datatype itself is email string
    
    #Here we are enforcing that CGPA should be greater than 0 and less than 10, we can even set the default value
    #Here we can even provide the description of the datatype, ( This will help LLMm to understand better obv, just like we have seen in typed-dict)
    cgpa: float = Field(gt=0, lt=10, default=5, description="This value shows the CGPA of the student in the semister")
```

Lets pass CGPA 12 and see if it throws an error

```python
new_student = {'age': "32", "email": 'abc@outlook.com', 'cgpa':12} 
student= Student(**new_student)
print(student)
```

Lets check the default value also

```python
new_student = {'age': "32", "email": 'abc@outlook.com'} 
student= Student(**new_student)
print(student)
```
